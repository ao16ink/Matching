```python
import cv2
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
# 入力画像とテンプレート画像をで取得
img = cv2.imread("/content/gdrive/My Drive/shawshank.jpg")
temp = cv2.imread("/content/gdrive/My Drive/shawshank_.jpg")
temp_ = cv2.imread("/content/gdrive/My Drive/shawshank__.jpg")

# グレースケール変換
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
temp = cv2.cvtColor(temp, cv2.COLOR_RGB2GRAY)
temp_ = cv2.cvtColor(temp_, cv2.COLOR_RGB2GRAY)

plt.subplot(131)
plt.imshow(gray)
plt.subplot(132)
plt.imshow(temp)
plt.subplot(133)
plt.imshow(temp_)
```

```python
# テンプレート画像の高さ・幅
h, w = temp.shape

# テンプレートマッチング（OpenCVで実装）
match = cv2.matchTemplate(gray, temp, cv2.TM_SQDIFF)
min_value, max_value, min_pt, max_pt = cv2.minMaxLoc(match)
pt = min_pt

# テンプレートマッチングの結果を出力
img_=img.copy()
cv2.rectangle(img_, (pt[0], pt[1]), (pt[0] + w, pt[1] + h), (0, 0, 200), 3)
img_ = cv2.cvtColor(img_, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10,10))
plt.subplot(131)
plt.imshow(temp)
plt.subplot(132)
plt.imshow(match)
plt.subplot(133)
plt.imshow(img_)
```
